```python
import numpy as np
import matplotlib.pyplot as plt
```

## Results for Fashion MNIST

```python
L_fmnist = np.loadtxt("./results/DA_evol_FMNIST_hierarchical_mmd", delimiter=",")
# L_fmnist = L_fmnist[None]
L_fmnist.shape
```

```python
m_fmnist = np.mean(L_fmnist, axis=0) * 100
std_fmnist = np.std(L_fmnist, axis=0) * 100
```

```python
absc = np.arange(0, 500001, 500000//100)

fig, ax = plt.subplots(figsize=(4,2)) #, gridspec_kw={"width_ratios":[1,2]})

plt.plot(absc, m_fmnist, label="MMDSW")
plt.fill_between(absc, m_fmnist-2*std_fmnist/np.sqrt(3), m_fmnist+2*std_fmnist/np.sqrt(3), alpha=0.5)

plt.plot(absc, 99.1 * np.ones((len(absc),)), label="OTDD")
plt.ylim(80, 101)

plt.xticks([0, 10**5, 2*10**5, 3*10**5, 4*10**5, 5*10**5], ["0", r"$10^5$", r"$2\cdot10^5$", r"$3\cdot10^5$", r"$4\cdot10^5$", r"$5\cdot10^5$"])
plt.ylabel("Accuracy", fontsize=13)
plt.xlabel("Iterations", fontsize=13)
plt.legend(fontsize=13)
plt.grid(True)

plt.show()
```

```python
absc = np.arange(0, 500001, 500000//100)[:50]
m_fmnist = m_fmnist[:50]
std_fmnist = std_fmnist[:50]

fig, ax = plt.subplots(figsize=(4,2)) #, gridspec_kw={"width_ratios":[1,2]})

plt.plot(absc, m_fmnist, label="MMDSW")
plt.fill_between(absc, m_fmnist-2*std_fmnist/np.sqrt(3), m_fmnist+2*std_fmnist/np.sqrt(3), alpha=0.5)

plt.plot(absc, 99.1 * np.ones((len(absc),)), label="OTDD (final)")
plt.ylim(80, 101)

# plt.xticks([0, 10**5, 2*10**5, 3*10**5, 4*10**5, 5*10**5], ["0", r"$10^5$", r"$2\cdot10^5$", r"$3\cdot10^5$", r"$4\cdot10^5$", r"$5\cdot10^5$"])

plt.xticks([0, 5*10**4, 10**5, 1.5*10**5, 2*10**5, 2.5*10**5], ["0", r"$5\cdot10^4$", r"$10^5$", r"$1.5\cdot10^5$", r"$2\cdot10^5$", r"$2.5\cdot10^5$"])

plt.ylabel("Accuracy", fontsize=13)
plt.xlabel("Iterations", fontsize=13)
plt.legend(fontsize=13)
plt.grid(True)

# plt.savefig("./evolution_DA_FMNIST.pdf", format="pdf", bbox_inches="tight")

plt.show()
```
